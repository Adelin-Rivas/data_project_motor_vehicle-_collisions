```python
import folium
import pandas as pd
```

```python
map_ = folium.Map(location=[40.693943, -73.985880], default_zoom_start=15)
```

```python
df2=pd.read_csv('NYPD_Motor_Vehicle_Collisions_-_Crashes_in_2018.csv')
df2.head()
value_for_head=df2[['ZIP CODE','NUMBER OF PEDESTRIANS INJURED']]
```

```python
df2.columns
```

```python
value_for_head=value_for_head[value_for_head['NUMBER OF PEDESTRIANS INJURED']>0]
value_for_head=value_for_head.dropna()
```

```python
value_for_head['ZIP CODE']=value_for_head['ZIP CODE'].astype(float).astype(int).astype(str)
```

```python
by_zipcode=value_for_head.groupby('ZIP CODE').sum()
```

```python
by_zipcode['zip']=by_zipcode.index
```

```python
by_zipcode
```

```python
map_
```

```python
zipcodes_agg=pd.read_csv('nyc-rolling-sales.csv')
zipcodes_agg=zipcodes_agg.dropna()
```

```python
zipcodes_agg['Zip']=zipcodes_agg['ZIP CODE'].astype(str)
zipcodes_agg.dtypes
```

```python
map_.choropleth(geo_data="nyc-zip-code-tabulation-areas-polygons.geojson",
             data=by_zipcode, # my dataset
             columns=['zip','NUMBER OF PEDESTRIANS INJURED'], # zip code is here for matching the geojson zipcode, sales price is the column that changes the color of zipcode areas
             key_on='properties.postalCode', # this path contains zipcodes in str type, this zipcodes should match with our ZIP CODE column
             fill_color='BuPu', fill_opacity=0.7, line_opacity=0.2,
             legend_name='Injuries by Zipcode')
```
