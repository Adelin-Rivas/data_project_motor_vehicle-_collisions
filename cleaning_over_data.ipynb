```python
crashes=pd.read_csv('NYC_Car_Crash_Clean')
```

```python
crashes['DATE']=pd.to_datetime(crashes['DATE'])
```

```python
date_filt=((crashes['DATE']>=pd.to_datetime('01/01/2016')) & (crashes['DATE']<pd.to_datetime('1/1/2019') )) 
```

```python
clean_value=crashes[date_filt]
clean_value
```
